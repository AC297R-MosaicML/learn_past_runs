```python
def visualize_images(images, label, count):
    fig, ax = plt.subplots(nrows=count, ncols=count, figsize=(20,20))
    class_i = 0
    for row in ax:
        for col in row:
            i = np.where(label == class_i)[0][0]
            col.imshow(images[i, :])
            col.set_xticks([])
            col.set_yticks([])
            col.set_title(class_to_label[label[i][0]])
            class_i += 1
    plt.show()
    
visualize_images(X_train, y_train, 10)
```

## CIFAR-10

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
unique, counts = np.unique(y_train, return_counts=True)
dict(zip(unique, counts))
```

```python
# source: https://huggingface.co/datasets/cifar10
class_to_label_10 = {0: 'airplane',
                     1: 'automobile',
                     2: 'bird',
                     3: 'cat',
                     4: 'deer',
                     5: 'dog',
                     6: 'frog',
                     7: 'horse',
                     8: 'ship',
                     9: 'truck'}
```

```python
def visualize_images10(images, label, count):
    fig, ax = plt.subplots(nrows=count // (count//2), ncols=count//2, figsize=(20,20))
    class_i = 0
    for row in ax:
        for col in row:
            i = np.where(label == class_i)[0][0]
            col.imshow(images[i, :])
            col.set_xticks([])
            col.set_yticks([])
            col.set_title(class_to_label_10[label[i][0]])
            class_i += 1
    plt.tight_layout()
    plt.show()
    
visualize_images10(X_train, y_train, 10)
```
